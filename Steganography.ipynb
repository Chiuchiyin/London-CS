```python
%matplotlib inline
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def find_hidden_message(image_array):
    dim1 = len(image_array)
    dim2 = len(image_array[0])
    dim3 = len(image_array[0][0])
    solution = np.zeros((4, 4, 3), dtype=np.uint8)
    i, j, k = 0,0,0
    while i < dim1:
        j, k = 0,0
        while j < dim2:
            k = 0
            while k <dim3:
                num = int(image_array[i, j, k])
                hexa = hex(num)
                last_digit = int(hexa[-1], 16)
                solution[i, j, k] = last_digit*16
                k +=1
            j +=1
        i +=1
    print('R ',solution[:, :, 0])
    print('G ',solution[:, :, 1])
    print('B ',solution[:, :, 2])
    img = Image.fromarray(solution, 'RGB')
    imgplot = plt.imshow(img)
```

```python
#image 1
image_1 = np.zeros((4, 4, 3), dtype=np.uint8)
image_1[:, :, 0] = ([[218, 250, 250, 218],[250, 255, 255, 255],[10, 255, 255, 10],[218, 10, 10, 218]])
image_1[:, :, 1] = ([[234, 250, 250, 234],[250, 240, 240, 240],[10, 208, 208, 10],[234, 154, 154, 234]])
image_1[:, :, 2] = ([[250, 42, 42, 250],[42, 32, 32, 32],[10, 112, 112, 10],[250, 154, 154, 250]])
img = Image.fromarray(image_1, 'RGB')
imgplot = plt.imshow(img)
```

```python
find_hidden_message(image_1)
```
